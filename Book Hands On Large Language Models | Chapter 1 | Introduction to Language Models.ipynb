# Phi-3

The first step is to load our model onto the GPU for faster inference. Note that we load the model and tokenizer separately (although that isn't always necessary).

```python
# from transformers import AutoModelForCausalLM, AutoTokenizer

# # Load model and tokenizer
# model = AutoModelForCausalLM.from_pretrained(
#     "microsoft/Phi-3-mini-4k-instruct",
#     device_map="cuda",
#     torch_dtype="auto",
#     trust_remote_code=True,
# )
# tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
# Import necessary classes from the transformers library
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",  # Specify the pre-trained model to load
    device_map="cuda",  # Assign model to the GPU for faster inference
    torch_dtype="auto",  # Automatically select the best data type for performance
    trust_remote_code=True,  # Allow loading custom model code if needed
)

# Load the corresponding tokenizer for the model
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
import warnings
warnings.filterwarnings('ignore')
```

Although we can now use the model and tokenizer directly, it's much easier to wrap it in a `pipeline` object:

```python
# from transformers import pipeline

# # Create a pipeline
# generator = pipeline(
#     "text-generation",
#     model=model,
#     tokenizer=tokenizer,
#     return_full_text=False,
#     max_new_tokens=500,
#     do_sample=False
# )
```

```python
# Import the pipeline function from the Transformers library
from transformers import pipeline

# Create a text generation pipeline using a pre-trained model and tokenizer
generator = pipeline(
    "text-generation",  # Specifies the task as text generation
    model=model,  # The pre-trained model to use (assumed to be defined earlier)
    tokenizer=tokenizer,  # The tokenizer associated with the model (assumed to be defined earlier)
    return_full_text=False,  # Ensures only the newly generated text is returned
    max_new_tokens=500,  # Limits the output to 500 newly generated tokens
    do_sample=False  # Disables sampling, making the generation deterministic
)

# This ensures the pipeline generates text in a controlled manner with a pre-defined model and tokenizer.
```

Finally, we create our prompt as a user and give it to the model:

```python
# # The prompt (user input / query)
# messages = [
#     {"role": "user", "content": "Create a funny joke about chickens."}
# ]

# # Generate output
# output = generator(messages)
# print(output[0]["generated_text"])
```

```python
# The prompt (user input / query)
messages = [
    {"role": "user", "content": "Create a funny joke about chickens."}
    # Defines the user input as a message in a structured format
]

# Generate output using the text generation pipeline
output = generator(messages)  # Passes the input message to the generator

# Print the generated joke
print(output[0]["generated_text"])  # Extracts and prints the generated text from the output
```

```python
# Print the raw output
print(output)
```
